```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# jupyter notebook ozel kod || plt.show() kullanmadan gormeyi saglar
%matplotlib inline
```

```python
# veri olusturalim
yasArray = [10, 20, 30, 40, 50, 60, 70, 75, 80, 80]
kiloArray = [20, 60, 80, 85, 70, 90, 95, 60, 55, 90]
```

```python
# numpy listeleri cok isimize yarayacagi icin bu verileri numpy array yapmakta fayda var
numpyYasArray = np.array(yasArray)
numpyKiloArray = np.array(kiloArray)
```

```python
# grafigi yazdirma 
#         x ekseni        y ekseni      renk
plt.plot(numpyYasArray, numpyKiloArray, 'g')
```

```python
# grafige labellar ekleyelim ve daha anlamli bir hal alsin || plt.xlabel() plt.ylabel() plt.title()
plt.plot(numpyYasArray, numpyKiloArray, 'r')
plt.title('Kilonun yasa gore degisim grafigi')
plt.xlabel('Yas')
plt.ylabel('Kilo')
# plt.show()   -----> %matplotlib inline kullandigim icin buna gerek kalmadi
```

```python
# yeni veri olusturalim ve baska islemler yapalim
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi1
```

```python
# baska bir veri olusturalim 
numpyDizisi2 = numpyDizisi1 ** 3
numpyDizisi2
```

```python
# grafigine bakalim ve cizgiyi ozellestirelim
plt.plot(numpyDizisi1, numpyDizisi2, 'g*')
```

```python
# cizgiyi ozellestirelim
plt.plot(numpyDizisi1, numpyDizisi2, 'go')
```

```python
# cizgiyi ozellestirelim
plt.plot(numpyDizisi1, numpyDizisi2, 'go-')
```

```python
# birden fazla cizgi grafik olsun istersek eger icinde plt.subplot()
# 1 sira olacak 2 grafik olacak 1.grafik
plt.subplot(1, 2, 1)
plt.plot(numpyDizisi1, numpyDizisi2, 'r*-')

# 1 sira olacak 2 grafik olacak 2.grafik
plt.subplot(1, 2, 2)
plt.plot(numpyDizisi2, numpyDizisi1, 'gv--')
```

## Ozellestirme

```python
# bos bir kanvas olarak dusunebiliriz bunu
benimFigure = plt.figure()

# 0 axes olan kanvasima deger ekleyelim (add_axes()). Bu kisim biraz karisik, arguman vercez (liste icinde olmali)
figureAxes = benimFigure.add_axes([0.1, 0.1, 0.6, 0.6])
figureAxes.plot(numpyDizisi2, numpyDizisi1, 'gv--')
figureAxes.set_xlabel('x ekseni veri ismi')
figureAxes.set_ylabel('y ekseni veri ismi')
figureAxes.set_title('grafik basligi')
```

```python
# yukarida bahsettigimiz add_axes icerigini daha iyi anlamak icin ornekler yapalim
# yeni bos figure olusturalim
figure2 = plt.figure()
#                   add_axes([grafik konum x, grafik konum y, grafik boy x, grafik boy y])
eksen1 = figure2.add_axes([0.1, 0.1, 0.7, 0.7])
eksen2 = figure2.add_axes([0.23, 0.4, 0.3, 0.3])

eksen1.plot(numpyDizisi1, numpyDizisi2, 'g')
eksen1.set_xlabel('x ekseni')
eksen1.set_ylabel('y ekseni')
eksen1.set_title('Ana Grafik Basligi')

eksen2.plot(numpyDizisi2, numpyDizisi1, 'b.')
eksen2.set_xlabel('x ekseni kucuk')
eksen2.set_ylabel('y ekseni kucuk')
eksen2.set_title('Kucuk Grafik Basligi')
```

### Gorsel Gelistirmeler

```python
# yeni figure olusturup gorselligi arttiralim
# ayni zamanda burada figure de belli degerler atanabilir. Ornegin dpi= x kalitesini belirler. alt+tab ile bunu rahatca gorebilirsin
yeniFigure = plt.figure(dpi = 150, facecolor='cyan')

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])

# burada label olarak yazdigimiz sey legend kullandigimiz zaman sol uste geldi
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 2, label = 'numpyDizisi ** 2')
yeniEksen.plot(numpyDizisi1, numpyDizisi2 ** 3, label = 'numpyDizisi ** 3')
# burada legend() bos da kullanilabilir loc = x seklinde lokasyon da atanabilir
yeniEksen.legend(loc = 10)

'''
NOT: Figure icinde yer alan dpi degerini daha sonra da ayarlayabiliriz. Ornegin bunu kaydetmek istiyoruz
Bunun ucun   yeniFigur.safefig('benimfig.png', dpi = 200)  seklinde pc ye kayit ederken de dpi ayarlanabilir
'''
```

### Ileri seviye gorsellestirme

```python
# cizgiler otelenebilir
#        x                y              renk               kalinlik           stil         isaretleme      isaretci boyu    isaretci rengi  
plt.plot(numpyDizisi1, numpyDizisi1 + 2, color = 'green', linewidth = 10, linestyle = '--', marker = 'o', markersize = 10  ,markerfacecolor = 'red')
plt.plot(numpyDizisi1, numpyDizisi1 + 3, color = 'red')
plt.plot(numpyDizisi1, numpyDizisi1 + 1, color = 'cyan', linewidth = 5, linestyle = '-.', marker = '*', markersize = 8  ,markerfacecolor = 'blue')
```
