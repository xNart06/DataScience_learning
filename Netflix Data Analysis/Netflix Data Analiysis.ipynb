```python
# kutuphaneler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# veri setini cekelim
netflix = pd.read_csv('netflix_titles.csv')
```

```python
# main dosya kaybolmamasi icin data degiskenine kopyaladik
data = netflix.copy()
data.shape
```

```python
# eksik verileri silelim
data = data.dropna()
data
```

```python
# tekrar bakalim data icerigine
data.shape
```

```python
# dizi ve film orani

# 'type' sütunundaki değerleri sayın
type_counts = netflix['type'].value_counts()

# Çubuk grafiğini çizin
sns.barplot(x=type_counts.index, y=type_counts.values)
plt.title('Tur')
plt.show()
```

```python
# rating tablosu
# 'rating' sütunundaki değerleri sayın
rating_counts = netflix['rating'].value_counts()

# Çubuk grafiğini çizin
plt.figure(figsize=(10, 6))
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.title('Rating')
plt.show()
```

```python
# Tur ve rating orani
plt.figure(figsize = (10,6))
sns.countplot(x='rating', hue = 'type', data = netflix)
plt.title('Tur ve rating orani')
plt.show()
```

```python
# pasta grafigi
labels = ['Movie', 'TV Show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0,1,2))
explode = [0,0.1]
plt.rcParams['figure.figsize'] = (9,9)
plt.pie(size, labels=labels, colors=colors, explode=explode, shadow = True)
plt.title('Tur Gorunumu', fontsize=25)
plt.legend()
```

```python
netflix['rating'].value_counts().plot.pie(autopct = '%1.1f%%', shadow = True)
```
