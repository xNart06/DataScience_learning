```python
# kutuphaneler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 2. ASAMA : EKSIK VERILER

```python
# isnull() ile kontrol edip sum() ile toplam eksik veriyi yazdiralim
df.isnull().sum()
```

```python
# eksik verileri seaborn ile goruntuleyelim
# once grafik icin ayarlamalari yapalim
sns.set_theme()
sns.set(rc={'figure.dpi':300, 'figure.figsize':(12,9)})
```

```python
# heatmap metodu ile eksik verileri goruntuleyelim
sns.heatmap(df.isnull(), cbar=False)
```

```python
# eksik verileri temizlemenin en kolay yolu silmektir. fakat Bu veri kaybina yol acacagi icin cok tavsiye edilmez.
# bu yuzden eksik veriler sutunun ortalamasi ya da medyani yazilir. Medyan aykiri degerelere karsi daha dayaniklidir
```

```python
# rating sutunun medyan degerini bulalim
rating_median = df['Rating'].median()
print(rating_median)
# simdi bu degeri fillna() ile eksik verilere atayalim
df['Rating'].fillna(rating_median, inplace=True)
```

```python
#diger sutunlardaki eksik veriler az oldugu icin silinebilir
df.dropna(inplace=True)

# simdi eksik verileri gorelim
df.isnull().sum().sum()
```

```python
# info() metodu ile dataset icindeki ozetlere bakalim
df.info()
```

```python
# simdi ise yukarda belirttigimiz sayisal olmasi gereken veri tiplerini sayisala cevirelim
# Once bu sutunun ozet istatistiklerine bakalim
df['Reviews'].describe()
```

```python
# veri tiplerini sayisala cevirelim
df['Reviews'] = df['Reviews'].astype('int64')

# bakalim duzelmis mi\ Duzelmis\ birler basamagina gore yuvarlayalim net gorelim
df['Reviews'].describe().round()
```

```python
# burada baktigimiz zaman standart sapma ortalamadan yuksek. Ortanca sayi ile arasinda cok fark var. Max sayi ise cok buyuk
# bu bilgi cebimizde bulunsun
```

```python
# size sutununa bakalim\ Tek degerleri gorelim
print('Toplan {} adet farkli deger var'.format(len(df['Size'].unique())))
df['Size'].unique()
```

```python
# onceklikler bunu int yapabilmek icin M ve K harfleri kalkmali
df['Size'].replace('M',' ', regex=True, inplace=True)
df['Size'].replace('k',' ', regex=True, inplace=True)

# kontrol edelim
df['Size'].unique()
```

```python
# fark edildigi uzere arada 1 tane stirg ifade var. Bu tur degisiminde bize sorun cikaracak
# bu veri yerine sutunun medyan degerini yazalim
# yazi haric degerleri sayisal yapip medyanini alalim
size_median = df[df['Size']!= 'Varies with device']['Size'].astype(float).median()
size_median
```

```python
# yazi yerine size_median degerini atayalim
df['Size'].replace('Varies with device', size_median, inplace = True)
```

```python
# numeric ifadeye cevirelim Not: df['Size'] df.Size yazilabilier
df.Size = pd.to_numeric(df.Size)
```

```python
# kontrol edelim
df.Size.head()
```

```python
# detaylarina bakalim
df.Size.describe().round()
```

```python
# Install sutunundaki degerleri duzenlemeye gecelim
df['Installs'].unique()
```

```python
# goruldugu uzere aralarda , ve sonlarda + ifadeleri var. bunlari kaldiralim
# bunu lambda fonksiyonu ile yapalim
df.Installs = df.Installs.apply(lambda x:x.replace('+',''))
df.Installs = df.Installs.apply(lambda x:x.replace(',',''))

#degerleri int degere cevirelim
df.Installs = df.Installs.apply(lambda x:int(x))
```

```python
# kontrol edelim
df['Installs'].unique()
```

```python
# price sutununu ele alalim simdi de
df['Price'].unique()
```

```python
# $ ve . ifadelerini temizleyelim
df.Price = df.Price.apply(lambda x:x.replace('$',''))
df.Price = df.Price.apply(lambda x:float(x))
```

```python
#kontrol edelim
df['Price'].unique()
```

```python
# genres sutununda kac cesit var bakalim
len(df['Genres'].unique())
```

```python
# ilk 10 satirina bakalim
df['Genres'].head(10)
```

```python
# goruldugu uzere bazi ifadelerde ; var. Bu alt kategoriyi temsil ediyor. Bu 2 ifadeyi ayiralim ve sadece Genres degerini alalim
# string ifade ile ugrastigimiz icin str yazdik. Hangi ifadeyi secmek istegimizi sonda belirttik
df['Genres'] = df['Genres'].str.split(';').str[0]
```

```python
# unique degerlere tekrar bakalim
len(df['Genres'].unique())
```

```python
# bu tek degerler nelermis bakalim
df['Genres'].unique()
```

```python
# sutundaki grup saylarina bakalim
df['Genres'].value_counts()
```

```python
# sondaki music&audio degeri tek oldugu icin ona music yazalim
df['Genres'].replace('Music & Audio', 'Music', inplace=True)
```

```python
# tekrar bakalim kodumuz dogru mu
df['Genres'].value_counts()
```

```python
# en son olarak Last update sutununu ele alalim
df['Last_Updated'].head()
```

```python
# object veri tipindeki bu veriyi datetime a cevirelim
df['Last_Updated'] = pd.to_datetime(df['Last_Updated'])
```

```python
# Bu islemler sonucunda dataset temizleme ve on isleme islemleri bitti. Son olarak veri setine bir daha bakalim
df.head()
```
