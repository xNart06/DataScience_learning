```python
dataFrame = dataFrame[dataFrame.year != 1970]
dataFrame.groupby('year').mean()['price']
```

```python
# transmission kolonu numeric olmadigi icin veriden kalkmali
dataFrame = dataFrame.drop('transmission', axis = 1)
```

```python
dataFrame.head()
```

```python
# ulasmak istedigimiz degeri 
y = dataFrame['price'].values
x = dataFrame.drop('price', axis = 1).values
```

```python
y
```

```python
x
```

```python
# train ve test icin kullanilacak fonksiyon kutuphanesini getiriyoruz
from sklearn.model_selection import train_test_split
```

```python
# train_test_split dokumantasyonda yer alan kodu yaziyoruz  test icin ayrilacak veri = test size
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state = 10)
```

```python
# kontrol edelim
len(x_train)
```

```python
len(x_test)
```

```python
# scaling islemi yapalim
from sklearn.preprocessing import MinMaxScaler
```

```python
# scaler nesnemiz
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
# artik tensorflow ile yapilacak islemlere gectik
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# sequential ile modeli    Dense ile katmanlari olusturuyoruz
```

```python
x_train.shape
```

```python
model = Sequential()

# noronlar
model.add(Dense(12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(12,activation='relu'))

# cikis
model.add(Dense(1))

# compiler kismini hazirliyoruz
model.compile(optimizer='adam', loss='mse')
```

```python
# model egitim kismina gelelim
model.fit(x = x_train, y = y_train, validation_data = (x_test, y_test), batch_size = 250 , epochs = 300)
```
